```python
import pandas as pd
from src.classifier import NaiveBayesTextClassifier
```

```python
df = pd.read_csv("data/spam_emails.csv")
texts = df["text"]
labels = df["label"]
```

```python
model = NaiveBayesTextClassifier()
model.train(texts, labels)
```

```python
predictions = model.predict([
    "Claim your free prize now!",
    "Let's meet at the library",
    "Hey, how are you doing today?"
])
print(predictions)
```

```python
accuracy = model.score(texts, labels)
print(f"Model accuracy: {accuracy}")
```
